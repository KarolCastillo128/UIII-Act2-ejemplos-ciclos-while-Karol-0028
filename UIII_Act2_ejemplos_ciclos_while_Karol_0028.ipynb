#### Ejemplos de while

```python
# 1. Ejemplo de while interactivo
# Este ejemplo cuenta desde 1 hasta el número que el usuario indique.

limite = int(input("¿Hasta qué número quieres contar? "))
contador = 1

while contador <= limite:
    print(f"Contador: {contador}")
    contador += 1
```

```python
# 3. Ejemplo de while con contraseña interactiva
# Este ejemplo solicita una contraseña hasta que sea correcta.

contraseña_correcta = "profe Eliseo Nava"
contraseña = ""

while contraseña != contraseña_correcta:
    contraseña = input("Ingresa la contraseña: ")

print("¡Acceso concedido!")
```

```python
# 4. Ejemplo de while con números pares interactivo
# Este ejemplo muestra los números pares hasta el límite indicado por el usuario.

limite = int(input("Ingresa el número límite: "))
numero = 2

while numero <= limite:
    print(f"Número par: {numero}")
    numero += 2
```

```python
# 5. Ejemplo de while con validación de acceso
# Este ejemplo verifica si una persona puede entrar a un evento.

edad = int(input("Ingresa tu edad: "))
boleto = input("¿Tienes boleto? (si/no): ")

while edad < 18 or boleto.lower() != "si":
    print("No puedes entrar al evento.")

    edad = int(input("Ingresa nuevamente tu edad: "))
    boleto = input("¿Tienes boleto? (si/no): ")

print("¡Puedes entrar al evento!")
```
